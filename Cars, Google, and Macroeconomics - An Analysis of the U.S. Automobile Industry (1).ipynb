##  Abstract

Using Economic Data compiled from various Government Libraries, this study provides a rough prediction of Annual Car Sales in the United States by utilizing OLS (Ordinary Least Squares Regression) and Logistic Regression. 3 of 10 features were ultimately found to be correlated with the target variable and used in the models. Data from Google Trends held the highest correction with the target variable, however the scope of the study was limited by the how recent Google began to track searches and make them available to the public. Nevertheless, this study demonstrated that Google Trends can be used effectively in analyzing and predicting Annual Car Sales in statistical models going forward. As further data in google searches is collected, this model should improve in accuracy to a large degree.

### 3.2 Workflow

What's the process you are going to follow to solve the problem based on the Problem-Solving Framework:

Problem
- Macroeconomic Indicators are not correlated with car sales high enough to provide a reasonably accurate predictive model for Annual Car Sales by themselves, it is necessary to implement modern data tools such as Google Trends to provide better features in order to successfully create one. 

Modeling Tools
- Modeling tools for this study include a dataset manually collected from various government and other public sources, OLS Regression, and Logistical Regression. 

Metrics
- The metrics used to evaluate the model are standard for OLS and Linear Regression and include correlation coefficients, P-Values, R Squared, and T scores.  

Work Flow
   - Data was gathered, cleaned, and shaped for use in model before being visualized in histograms and plots.
   - Histograms revealed a high degree of skewness in most features.
   - The Data was normalized to correct skewness using a logarithmic function. A Lambda function applied to all variables that required correction.
   - "CarSales" was chosen to be the target variable and was normalized with a logarithmic function as well. 
   - Once the data was normalized, a correlation matrix was created in order to provide further insight into which features were correlated with the target variable.
   - Ordinary Least Squares Regression was then used in order to isolate the influences of correlated regressors on the dependent variable. 
   - The OLS model was totally successful with a Coefficient of Determination equaling 0.973. However it quickly became apparent that the model was actually overfitted as opposed to unbelievably accurate, and the process for feature selection to refine the model was begun. 
   - In order to eliminate "noise" that was causing linear regression to confuse random values with correlated ones, features without a correlation rate of at least 0.6 with the target variable were dropped. 

Final Model
- Multilinear regression was used as the final model once all the features were engineered and selected in order to incorporate every useful feature into the model and receive better results. 
     
Model Implementation
- The 3 features chosen as independent variables were:
 - Average Yearly Value of S&P 500 Index
 - Average Yearly Unemployment Rate
 - Average Yearly Google Trends Score

- Projected 2021 values from analysts of the S&P500 and Unemployment Rate was then used in tandem with the 2020 Google Trends Score to predict the number of new cars that will be sold in 2021. 

Evaluate Model
- The total number of passenger vehicles that will be sold in 2021 is projected to be 15,880,468, higher than the 2020 value of 14,665,787, an 8.28% increase.

```python
# Normalizing and Cleaning Data for Final Model
Cartrans2 = Cars.apply(lambda x: np.log10(x) if np.issubdtype(x.dtype, np.number) else x)
Cartrans2 = Cartrans2.replace([np.inf, -np.inf], np.nan)
Cartrans2 = Cartrans2.fillna(0)
```

```python
LR=LinearRegression()
LR.fit(X_train,y_train)
```

```python
#Fitting Model
LR.fit(X_train, y_train)
y_predLR = LR.predict(X_test)
```

```python
#Finalizing Model
r2_score(y_test, y_predLR)
```
